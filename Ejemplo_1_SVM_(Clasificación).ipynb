## Graficar los datos transformado por la funci√≥n kernel

```python
from sklearn.kernel_approximation import Nystroem
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Aproximamos el espacio transformado usando un kernel RBF
feature_map = Nystroem(kernel='rbf', gamma=1, n_components=3)
X_transformed = feature_map.fit_transform(X)  # Transformamos los datos a 3D
```

```python
# Entrenar SVM en el espacio transformado
svm = SVC(kernel="linear")
svm.fit(X_transformed, y)

# Obtener los coeficientes del hiperplano
w = svm.coef_[0]  # Coeficientes del hiperplano
b = svm.intercept_[0]  # T√©rmino de sesgo
```

```python
w, b
```

```python
# Crear una malla para visualizar el hiperplano
x_range = np.linspace(X_transformed[:, 0].min(), X_transformed[:, 0].max(), 30)
y_range = np.linspace(X_transformed[:, 1].min(), X_transformed[:, 1].max(), 30)
X_grid, Y_grid = np.meshgrid(x_range, y_range)

# Ecuaci√≥n del hiperplano: w1*x + w2*y + w3*z + b = 0 -> despejamos z
Z_grid = (-w[0] * X_grid - w[1] * Y_grid - b) / w[2]

# Visualizaci√≥n en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_transformed[:, 0], X_transformed[:, 1], X_transformed[:, 2], c=y, cmap='coolwarm', edgecolor='k')
ax.plot_surface(X_grid, Y_grid, Z_grid, alpha=0.3, color='gray')  # Hiperplano de separaci√≥n

ax.set_xlabel('Feature 1 (Transformado)')
ax.set_ylabel('Feature 2 (Transformado)')
ax.set_zlabel('Feature 3 (Nueva dimensi√≥n)')
plt.title("Datos en el espacio del kernel con hiperplano SVM")
plt.show()
```
